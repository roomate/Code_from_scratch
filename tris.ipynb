# Sort algorithms

I present in this notebook four very classical sorting algorithms: Insertion sort, Quick sort, Merge sort and Heap sort.

```python
import numpy as np
import matplotlib.pyplot as plt
from functools import wraps

def check_sorted(L: list):
    """
    Check if a list is correctly sorted.
    """
    for i in range(1, len(L)):
        if L[i] < L[i - 1]:
            return False
    return True
```

### Benchmark the time

```python
def timewrapper(f):
    @wraps(f)
    def time_slot(*args, **kwargs):
        import time
        t0 = time.time()
        res = f(*args, **kwargs)
        t1 = time.time()
        return t1 - t0, res
    return time_slot
```

## Insertion Sort

The insertion sort is the most intuitive one, yet not the most efficient. It consists in browsing the list (say from left to right), and swap with its left neighbours while it is larger.

In the best case, the time complexity is $\mathcal{O}(N)$. It happens when the list is already sorted. On the exact opposite, when the list is sorted in decreasing order, this is the worst case, and the time complexity scales as $\mathcal{O}(N^2)$. The average complexity is $\mathcal{O}(N^2)$, so this algorithm is seldomly chosen, except when one knows that the list is 'almost' sorted. Memory complexity is always $\mathcal{O}(N)$ though, so can be advantageous when not a lot of memory is available.

```python
def swap(L,i,j):
    "Swap i-th and j-th elements of the list L"
    a = L[i]
    L[i] = L[j]
    L[j] = a
    return L

@timewrapper
def insert_sort(L):
    "Sort the list L"
    for i in range(1,len(L)):
        j = i - 1
        while j>= 0 and (L[j] > L[i]):
            L = swap(L, i, j)
            i = j
            j -= 1
    return L

N = [10, 100, 200, 500, 700, 1000, 2000, 3000, 4000, 5000, 7000, 10000, 12500, 15000, 20000, 25000, 30000]
T = []
for n in N:
    L = [np.random.randint(n) for _ in range(n)]
    time_mes, L = insert_sort(L)
    if not check_sorted(L):
        break
    T.append(time_mes)

plt.plot(N,T)
plt.show()
```

# Quick sort

Quick sort is a recursive algorithm. The main mecanism is to divide a list in two by the choice of a pivot element `x`, often taken as the list's very first element by default. Other choices are possible though. All elements smaller or equal than the pivot are put into the first list `L1`, while the bigger elements are put into the second list `L2`. The algorithm can then be called on these two lists afterwards. Once both are sorted, one can concatenate them, with first `L1`, then `x` and finally `L2`.

On average, time complexity is $\mathcal{O}(N\log(N))$, which makes it much more efficient than insertion sort.

```python
import numpy as np

@timewrapper
def quick_sort(L):
    if len(L) == 1 or len(L) == 0:
        return L
    pivot = L[0]
    L1 = []
    L2 = []
    for l in L[1:]:
        if l < pivot:
            L1 = np.append(L1,l)
        else:
            L2 = np.append(L2,l)
    _, L1 = quick_sort(L1)
    _, L2 = quick_sort(L2)
    return np.concatenate((L1, [pivot], L2), axis = 0)

N = [10, 100, 200, 500, 700, 1000, 2000, 3000, 4000, 5000, 7000, 10000, 12500, 15000, 20000, 25000, 30000, 40000, 45000,50000]
T = []
for n in N:
    L = [np.random.randint(n) for _ in range(n)]
    time_mes, L = quick_sort(L)
    if not check_sorted(L):
        break
    T.append(time_mes)

plt.plot(N,T)
plt.show()
```

# Merge sort

Merge sort is a recursive algorithm. It also starts by dividing a list in two parts, taking for example the first and last $N/2$ entries. This list is recursively divided in two while one gets lists of two elements, which are trivial to sort. Then, making the return trip, one just needs to make one big sorted list from two smaller sorted lists. 
Two functions needs to be implemented for this algorithm. The function `merge_sort` takes two sorted lists as input, and return the sorted concatenation. `recur` function recursively split the list in two, and calls `merge_sort` on them.

Time average complexity is $\mathcal{O}(N\log(N))$.

```python
def merge_sort(L1: list, L2: list):
    i1, i2 = 0, 0
    L = []
    while i1 != len(L1) and i2 != len(L2):
        a = L1[i1]
        b = L2[i2]
        if L1[i1] < L2[i2]:
            L.append(a)
            i1 += 1
        else:
            L.append(b)
            i2 += 1
    if i1 == len(L1):
        J = L2[i2:]
    else:
        J = L1[i1:]
    return np.concatenate((L,J), axis = 0)

@timewrapper
def recur(L):
    if len(L) == 1 or len(L) == 0:
        return L
    split_1, split_2 = np.split(L, indices_or_sections=[len(L)//2])
    return merge_sort(recur(split_1)[1], recur(split_2)[1])

N = [10, 100, 200, 500, 700, 1000, 2000, 3000, 4000, 5000, 7000, 10000, 12500, 15000, 20000, 25000, 30000, 35000, 40000, 45000, 50000, 100000]
T = []
for n in N:
    L = [np.random.randint(n) for _ in range(n)]
    time_mes, L = recur(L)
    if not check_sorted(L):
        print("Warning")
        break
    T.append(time_mes)
plt.plot(N,T)
plt.show()
```

# Heap Sort

The data structure <code>heap</code> can be represented in two manners: 
1. A list where elements $i$ has two children at $2i$ and $2i+1$, whose value is inferior to that of the parent. Note that for a list of $n$ elements, only the $(n - 1)/2$-th first elements have children.
2. A binary tree.

If we take the point of view of a binary tree, when we heapify the tree, we are guaranteed the largest element is located at the root of the binary tree.

A first function <code>is_heap</code> simply checks that the list is actually heapified.
A second function <code>make_it_heap</code> heapify the tree sequentially, starting from the root and going up, one node at a time. 
Once the tree is heapified, we are guaranted the root's value is the biggest. Then, you can exchange the root and last node of the list, and prune the tree by removing the last element of the list. You heapify the tree once again to have the biggest elements at the root's node and repeat again. You continue until there remains only one element in the list.

Both worst and best time complexity scale as $\mathcal{O}(N\log(N))$.

```python
def is_heap(L: list, order = "decrease"):
    n = len(L)//2
    if order == "decrease":
        for i in range(n):
            if L[2*i + 1] > L[i]:
                return False
            try:
                if L[2*i + 2] > L[i]:
                    return False
            except IndexError:
                pass
    elif order == 'increase':
        for i in range(n):
            if L[2*i + 1] < L[i]:
                return False
            try:
                if L[2*i + 2] < L[i]:
                    return False
            except IndexError:
                pass
    return True

def make_it_heap(heap: list, length: int):
    idx = length - 1
    start = (idx - 1)//2
    while start >= 0:
        browse_tree(heap, start, length)
        start -= 1

def browse_tree(heap: list, root: int, length: int):
    while 2*root < length: #While there is a left child to check, the root is potentially smaller that one of its children.
                                #The goal is to make the children go down one stage at a time.
        child = 2*root

        if child + 1 < length and heap[child + 1] > heap[child]:
            child += 1

        if heap[child] > heap[root]:
            heap[child], heap[root] = heap[root], heap[child]
            root = child
        else:
            return #Because the sub trees are supposed to be heapified, no need to go further

@timewrapper
def heap_sort(heap: list, length: int | None) -> None:
    if length is None:
        length = len(heap)

    make_it_heap(heap, length)
    while length > 1:
        length -= 1
        heap[0], heap[length] = heap[length], heap[0] #Take the largest element (at root node after heapification), and put it into the deeper node.

        make_it_heap(heap, length)

N = [10, 100, 200, 500, 700, 1000, 2000, 3000, 4000, 5000, 7000, 10000, 12500, 15000, 20000, 25000, 30000]
T = []
for n in N:
    L = [np.random.randint(n) for _ in range(n)]
    time_mes = heap_sort(L, len(L))
    if not check_sorted(L):
        break
    T.append(time_mes)

plt.plot(N,T)
plt.show()
```
